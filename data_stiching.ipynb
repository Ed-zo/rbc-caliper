```python
import os
import pandas as pd
import re

folder = 'krvinky/'
output_folder = 'output/'

# Get all files from folder
files = os.listdir(folder)
p = re.compile('(.*)-(\d*).tif.measured.csv')

files = [ p.match(s) for s in files if p.match(s) ]

all_count = 0
grouped_files = {}
for f in files:
    num = f.groups()[0]
    if num not in grouped_files:
        grouped_files[num] = set()

    grouped_files[num].add(f.string)

all = None
for name, files in grouped_files.items():
    data = None
    for file in files:
        df = pd.read_csv(folder + file, index_col=0)
        df = df.round(3)
        
        if data is None:
            data = df
        else:
            data = pd.concat([data, df], axis=0)


    data['id'] = int(name[2:])
    if all is None:
        all = data
    else:
        all = pd.concat([all, data], axis=0)

    all_count += len(data)
    if len(data) < 100:
        print(f'!!!! Not enough data for {name}. Only got {len(data)} rows !!!!')
    else:
        print(f'Got {len(data)} rows for {name}')

    data.to_csv(output_folder + name + '.csv', index=False)

all.to_csv(output_folder + 'all.csv', index=False)
print(f'Got {all_count} rows in total from {len(grouped_files)} files.')
```
